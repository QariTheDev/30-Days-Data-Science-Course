## Case Study


```python
#importing Data
data=pd.read_csv('titanic.csv')
#checking data
data.head()
```

```python
data.info()
```

```python
#describe
data.describe()
```

```python
#droping columns
data=data.drop(['parch'],axis=1)
data.head()
```

```python
#droping rows
data.head()
```

```python
#mean
data.mean()
```

```python
#groupby
data.groupby(['sex','pclass']).mean()
```

```python
#value counts
data.value_counts('survived')
```

```python

#specific data
#under 18 
under_18=data[data['age']<18]
under_18.groupby(['sex','pclass']).mean()
```
