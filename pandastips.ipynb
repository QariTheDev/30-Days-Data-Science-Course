# Pandas Tips and Tricks

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#importing data
kashti=sns.load_dataset('titanic')
kashti.head(10)
```

```python
#finding
kashti.shape
```

### 1. Changing Column names

```python
kashti.rename(columns={'survived':'zinda'},inplace=True)
kashti.head()
```

### 2. Reverse Row Order

```python
kashti.loc[::-1]
```

```python
kashti.loc[::-1].reset_index(drop=True)
```

### 3. Reverse Column Order


```python
kashti.loc[:,::-1]
```

### 4. Changing Data Type

```python
kashti=kashti.astype({'alone':'int64'})
kashti.head()
```

### 5. Finding Columns with dtype

```python
kashti.dtypes
```

```python
kashti.select_dtypes(['object','category']).head()
```

### 6. Split Data

```python
data40=kashti.sample(frac=0.40)
data40.shape
```

```python
data60=kashti.drop(data40.index)
data60.shape
```

```python
data60
```

### 7. Joining Data

```python
kashti1=pd.concat([data40,data60])
kashti1
```

### 8. Filtering Data

```python
femaledata=kashti[(kashti['sex']=='female') &(kashti['age']>30) ]
femaledata
```

```python
maledata=kashti[(kashti['sex']=='male')&(kashti['age']>25)&(kashti['pclass']==1)]
maledata
```

```python
child=kashti[(kashti.age<18) & (kashti.sex=='male')]
child.shape
```

### 9. Group By

```python
kashti.groupby('sex').mean()
```

```python
kashti.groupby('sex').count()
```

```python
len(kashti.groupby('age'))
```

```python
kashti.groupby(['sex','pclass','alive']).mean()
```
